# Dataset D3

## Importing libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as tkr
```

## Data reading and formating

```python
data = pd.read_csv('../summary_data/D3_TCRA_simulated.csv')
data['total_corrections']=data['Base - TP']+ data['Base - FP']
```

## Defining color dictionary

```python
color_dict=dict({'Bfc':'purple','Bless':'orange','Coral':'brown','Fiona':'gray','Lighter':'pink','Musket':'blue','Pollux':'yellow','Racer':'green','Reckoner':'red','Sga':'black'})
```

## Selecting best kmer for each tool

```python
data_best = data.loc[data.groupby(["Tool", "Coverage"])["Base Gain"].idxmax()]  
```

<br>
<br>

# Figure S12a
The effect of kmer size on the accuracy of the error correction tools across for TCR-Seq data for coverage of 1x. <br> The reads were simulated from TCRA transcripts.

```python
sns.set_style("white")
sns.set_context("poster")

g=sns.barplot(x="Kmer Size", 
              y="Base Gain",
              data = data[(data['Coverage'] ==1) & (data['Tool'] !='Fiona') & (data['Tool'] !='Racer')],
              hue='Tool', 
              palette=color_dict)

g.set(xlabel='K-mer size', ylabel='Gain')

plt.ylim( 0, 1)
plt.title("Coverage 1x", fontsize=30)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)


fig = plt.gcf()
fig.set_size_inches(16, 10)

sns.despine()

plt.savefig("../figures/D3_TCRA_simulated/FigS12a_gain_vs_kmer_cov1.png", bbox_inches='tight')
plt.savefig("../figures/D3_TCRA_simulated/FigS12a_gain_vs_kmer_cov1.pdf", bbox_inches='tight')
```

# Figure S14
Heatmap depicting the gain across various coverage settings of TCR-seq.
<br>Each row corresponds to an error correction tool, and each column corresponds to a dataset with a given coverage.
<br> For each tool, the best k-mer size was selected.


```python
result = data_best.pivot(index='Tool', columns='Coverage', values='Base Gain')\
    .sort_values(by="Tool", ascending=False)

sns.set_style("white")
sns.set_context("talk")

fig, ax = plt.subplots(figsize=(8,5))
g=sns.heatmap(result, 
              annot=True, 
              cmap='coolwarm',
              center=0,
              linewidths=.5,
              annot_kws={'size':15},
              fmt=".2f",
              vmin=-1, 
              vmax=1)

g.set(xlabel='Coverage', ylabel='Gain')
g.set_ylim(0, 10)

sns.despine()

plt.savefig("../figures/D3_TCRA_simulated/FigS14_heatmap_humna_gain.png", bbox_inches="tight")
plt.savefig("../figures/D3_TCRA_simulated/FigS14_heatmap_human_gain.pdf", bbox_inches="tight")
```

# Figure S17a
Trimming Efficiency vs. Trim Portion for TCR-Seq simulated data for coverage 1x.
<br> For each tool, the best k-mer size was selected.


```python
sns.set_style("white")
sns.set_context("talk")

g=sns.lmplot(data=data_best[(data_best['Coverage'] == 1)] , 
             x='Trim Percent', 
             y='Trim Efficiency', 
             hue='Tool', 
             palette=color_dict, 
             fit_reg=False,
             aspect=1.3,
             scatter_kws={"s": 150})

g.set(xlabel='Trim Portion', ylabel='Trim Efficiency')

plt.ylim(-0.1, 1.1)
plt.xlim(-0.1, 1.1)
plt.title("Coverage 1x", fontsize=20)

sns.despine()

plt.savefig("../figures/D3_TCRA_simulated/FigS17a_efficiency_vs_percent_cov1.png", bbox_inches="tight")
plt.savefig("../figures/D3_TCRA_simulated/FigS17a_efficiency_vs_percent_cov1.pdf", bbox_inches="tight")
```
