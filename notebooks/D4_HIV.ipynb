## Data reading and formating - D4 HIV mixture

```python
data = pd.read_csv('../summary_data/D4_HIV_summary.csv')
```

## Defining color dictionary

```python
color_dict=dict({'Bfc':'purple','Bless':'orange','Coral':'brown','Fiona':'gray','Lighter':'pink','Musket':'blue','Pollux':'yellow','Racer':'green','Reckoner':'red','Sga':'black'})
```

## Selecting best kmer for each tool

```python
data_best = data.loc[data.groupby(["Tool"])["Base Gain"].idxmax()]  
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("white")
sns.set_context("poster")

g= sns.barplot(y="Tool", 
              x="Base Gain",
              data = data_best.sort_values(by="Base Gain", ascending=True), 
              palette=color_dict,
             errwidth=3)


#x1.set_xlim(-13, -9)
#x2.set_xlim(-.9, 0.5)

g.set(xlabel='Gain', ylabel='Tool')
g = sns.despine()

#plt.xlim(-1, 1)
fig = plt.gcf()
fig.set_size_inches(10, 6)
#x.text(-1.20, 12,'Gain', fontsize=22)
plt.savefig("../figures/D4_HIV/Fig4a_gain.png", bbox_inches='tight')
plt.savefig("../figures/D4_HIV/Fig4a_gain.pdf", bbox_inches='tight')
```

## Figure 4a
Error correction method evaluation based on viral sequencing data (D4 dataset). 
<br> Error-free reads for gold standard were obtained using UMI-based protocol. 
<br> Datasets with various coverage were obtained by downsampling the original data.
<br> For each tool, the best k-mer size was selected.

```python
g=sns.set_style("white")
g=sns.set_context("talk")


g=sns.lmplot(data = data_best,
             x='Base Precision', 
             y='Base Gain', 
             hue='Tool', 
             palette=color_dict, 
             fit_reg=False,
             aspect=1.3,
             scatter_kws={"s": 150})

g.set(xlabel='Precision', ylabel='Gain')
#plt.ylim(-0.1, 1.1)
#plt.xlim(-0.1, 1.1)

g=sns.despine()
plt.savefig("../figures/D4_HIV/Fig4a_Precision_vs_Gain.png")
plt.savefig("../figures/D4_HIV/Fig4a_Precision_vs_Gain.pdf")
```

## Figure 4b
Scatter plot depicting the sensitivity (x-axis) and precision (y-axis) of each tool when applied to D5 HIV mixture dataset.
<br>
For each tool, the best k-mer size was selected.


```python

g=sns.set_style("white")
g=sns.set_context("talk")


g=sns.lmplot(data = data_best,
             x='Base - TP', 
             y='Base - FP', 
             hue='Tool', 
             palette=color_dict, 
             fit_reg=False,
             aspect=1.3,
             scatter_kws={"s": 150})\

for ax in g.axes.flatten():
    ax.set_xticklabels(ax.get_xticklabels(), fontsize=17)
    ax.xaxis.set_major_formatter(tkr.FuncFormatter(lambda x, p: "{:.0f}k".format(x/1000)))
    ax.yaxis.set_major_formatter(tkr.FuncFormatter(lambda x, p: "{:.0f}k".format(x/1000)))
    ax.yaxis.set_major_locator(tkr.MultipleLocator(100000))
    ax.xaxis.set_major_locator(tkr.MultipleLocator(75000))

g.set(xlabel='TP', ylabel='FP')
#plt.ylim(-0.1, 1.1)
#plt.xlim(-0.1, 1.1)

g=sns.despine()
plt.savefig("../figures/D4_HIV/Fig4b_Precision_vs_Sensitivity.png")
plt.savefig("../figures/D4_HIV/Fig4b_Precision_vs_Sensitivity.pdf")
```
